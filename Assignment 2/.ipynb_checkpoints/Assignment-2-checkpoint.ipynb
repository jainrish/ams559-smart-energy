```python
n = 672
x=cp.Variable((n))
objective=0
x[0]=0
for i in range(1,n):
    objective+= 0.4*x[i]

constraints = [x[:]>0]
prob = cp.Problem(cp.Minimize(objective), constraints)
result = prob.solve()
print(x)
```
